# task 1

```python
import math

class MinimaxSolver:
    def __init__(self):
        self.tree = {
            'Root': ['N1', 'N2'],
            'N1': ['N3', 'N4'],
            'N2': ['N5', 'N6'],
            'N3': [4, 4],
            'N4': [7, 2],
            'N5': [5, 1],
            'N6': [8, 3]
        }
        self.visited_order = []
        self.node_values = {}

    def solve(self, node, depth, is_max, limit=None):
        self.visited_order.append(node)
        
        if isinstance(node, int):
            return node
            
        if limit is not None and depth == limit:
            self.node_values[node] = 0
            return 0

        if is_max:
            best = -math.inf
            for child in self.tree[node]:
                val = self.solve(child, depth + 1, False, limit)
                best = max(best, val)
            self.node_values[node] = best
            return best
        else:
            best = math.inf
            for child in self.tree[node]:
                val = self.solve(child, depth + 1, True, limit)
                best = min(best, val)
            self.node_values[node] = best
            return best

solver = MinimaxSolver()
root_val = solver.solve('Root', 0, True)

print("Minimax Values:", solver.node_values)
print("Order of Visitation:", solver.visited_order)

solver_limited = MinimaxSolver()
root_val_limited = solver_limited.solve('Root', 0, True, limit=2)
print("Depth-Limited (2) Values:", solver_limited.node_values)
```

# task 2

```python
import math

class AlphaBetaSolver:
    def __init__(self):
        self.tree = {
            'Root': ['N1', 'N2'],
            'N1': ['N3', 'N4'],
            'N2': ['N5', 'N6'],
            'N3': [4, 4],
            'N4': [7, 2],
            'N5': [5, 1],
            'N6': [8, 3]
        }
        self.pruned = []
        self.node_values = {}

    def solve(self, node, depth, alpha, beta, is_max):
        if isinstance(node, int):
            return node

        if is_max:
            best = -math.inf
            for child in self.tree[node]:
                val = self.solve(child, depth + 1, alpha, beta, False)
                best = max(best, val)
                alpha = max(alpha, best)
                if beta <= alpha:
                    idx = self.tree[node].index(child)
                    self.pruned.extend(self.tree[node][idx+1:])
                    break
            self.node_values[node] = best
            return best
        else:
            best = math.inf
            for child in self.tree[node]:
                val = self.solve(child, depth + 1, alpha, beta, True)
                best = min(best, val)
                beta = min(beta, best)
                if beta <= alpha:
                    idx = self.tree[node].index(child)
                    self.pruned.extend(self.tree[node][idx+1:])
                    break
            self.node_values[node] = best
            return best

ab_solver = AlphaBetaSolver()
result = ab_solver.solve('Root', 0, -math.inf, math.inf, True)
print("Minimax Values with Pruning:", ab_solver.node_values)
print("Pruned Nodes/Branches:", ab_solver.pruned)
```

# task 3

```python
import math

tree = {
    'Root': ['N1', 'N2', 'N7'],
    'N1': ['N3', 'N4'],
    'N2': ['N5', 'N6'],
    'N7': [10, 12],
    'N3': [4, 4],
    'N4': [7, 2],
    'N5': [5, 1],
    'N6': [8, 3]
}

def minimax_path(node, is_max):
    if isinstance(node, int):
        return node, [node]
    
    if is_max:
        best_val = -math.inf
        best_path = []
        for child in tree[node]:
            val, path = minimax_path(child, False)
            if val > best_val:
                best_val = val
                best_path = [node] + path
        return best_val, best_path
    else:
        best_val = math.inf
        best_path = []
        for child in tree[node]:
            val, path = minimax_path(child, True)
            if val < best_val:
                best_val = val
                best_path = [node] + path
        return best_val, best_path

val, path = minimax_path('Root', True)
print("Updated Root Value:", val)
print("Optimal Path:", " -> ".join(map(str, path)))
```

# task 4

```python
from ortools.sat.python import cp_model

model = cp_model.CpModel()
a = model.NewIntVar(0, 3, 'A')
b = model.NewIntVar(0, 3, 'B')
c = model.NewIntVar(0, 3, 'C')

model.Add(a != b)
model.Add(b != c)
model.Add(a + b <= 4)

solver = cp_model.CpSolver()
status = solver.Solve(model)

if status == cp_model.OPTIMAL or status == cp_model.FEASIBLE:
    print(f"A = {solver.Value(a)}, B = {solver.Value(b)}, C = {solver.Value(c)}")
```

# task 5

```python
from ortools.sat.python import cp_model

class VarSolutionPrinter(cp_model.CpSolverSolutionCallback):
    def __init__(self, variables):
        cp_model.CpSolverSolutionCallback.__init__(self)
        self.__variables = variables

    def on_solution_callback(self):
        print(f"A={self.Value(self.__variables[0])}, B={self.Value(self.__variables[1])}, C={self.Value(self.__variables[2])}")

model = cp_model.CpModel()
a = model.NewIntVar(0, 3, 'A')
b = model.NewIntVar(0, 3, 'B')
c = model.NewIntVar(0, 3, 'C')

model.Add(a != b)
model.Add(b != c)
model.Add(a + b <= 4)

solver = cp_model.CpSolver()
solver.parameters.enumerate_all_solutions = True
solution_printer = VarSolutionPrinter([a, b, c])
solver.Solve(model, solution_printer)
```

# task 6

```python
from ortools.sat.python import cp_model

model = cp_model.CpModel()
x = model.NewIntVar(0, 20, 'x')
y = model.NewIntVar(0, 20, 'y')
z = model.NewIntVar(0, 20, 'z')

model.Add(x + 2*y + z <= 20)
model.Add(3*x + y <= 18)

model.Maximize(4*x + 2*y + z)

solver = cp_model.CpSolver()
status = solver.Solve(model)

if status == cp_model.OPTIMAL:
    print("Optimal Value:", solver.ObjectiveValue())
    print(f"x: {solver.Value(x)}, y: {solver.Value(y)}, z: {solver.Value(z)}")
```

# task 7

```python
from ortools.sat.python import cp_model

model = cp_model.CpModel()
n = 4
queens = [model.NewIntVar(0, n - 1, f'q{i}') for i in range(n)]

model.AddAllDifferent(queens)
model.AddAllDifferent([queens[i] + i for i in range(n)])
model.AddAllDifferent([queens[i] - i for i in range(n)])

solver = cp_model.CpSolver()
status = solver.Solve(model)

if status == cp_model.OPTIMAL or status == cp_model.FEASIBLE:
    for i in range(n):
        row = ["_"] * n
        row[solver.Value(queens[i])] = "Q"
        print("".join(row))
```
