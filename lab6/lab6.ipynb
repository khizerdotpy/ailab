# q1

```python
import random

def f(x):
    return -x**2 + 6*x

def hill_climbing(x_range):
    current_x = random.randint(x_range[0], x_range[1])
    current_f = f(current_x)
    
    print(f"Initial value of x: {current_x}")
    print(f"Value of f(x) at each step:")
    print(f"Step 0: x = {current_x}, f(x) = {current_f}")

    step = 1
    while True:
        neighbors = []
        if current_x + 1 <= x_range[1]:
            neighbors.append(current_x + 1)
        if current_x - 1 >= x_range[0]:
            neighbors.append(current_x - 1)
        
        next_x = current_x
        next_f = current_f
        
        for neighbor in neighbors:
            if f(neighbor) > next_f:
                next_x = neighbor
                next_f = f(neighbor)
        
        if next_x == current_x:
            break
            
        current_x = next_x
        current_f = next_f
        print(f"Step {step}: x = {current_x}, f(x) = {current_f}")
        step += 1
        
    return current_x, current_f

optimal_x, optimal_f = hill_climbing((0, 6))
print(f"\nFinal optimal value: x = {optimal_x}, f(x) = {optimal_f}")
```

# q2

```python
def heuristic(n, goal=20):
    return abs(goal - n)

def beam_search(start=1, goal=20, k=2):
    current_level = [([start], 0)]
    level_count = 0
    
    print(f"Goal: {goal}, Beam Width: {k}")
    
    while current_level:
        print(f"Level {level_count} explored states: {[path[-1] for path, h in current_level]}")
        
        for path, h in current_level:
            if path[-1] == goal:
                return path
        
        candidates = []
        for path, _ in current_level:
            current_n = path[-1]
            
            successors = [current_n + 2, current_n + 3, current_n * 2]
            
            for s in successors:
                if s <= goal + 3:
                    new_path = path + [s]
                    candidates.append((new_path, heuristic(s, goal)))
        
        candidates.sort(key=lambda x: x[1])
        current_level = candidates[:k]
        level_count += 1
        
    return None

final_path = beam_search(start=1, goal=20, k=2)

if final_path:
    print(f"\nFinal path to reach 20: {' -> '.join(map(str, final_path))}")
else:
    print("\nGoal not reached.")
```

# q3

```python
import random

def fitness_function(x):
    return x**2 + 2*x

def binary_to_int(chromosome):
    return int("".join(map(str, chromosome)), 2)

def genetic_algorithm(pop_size=4, generations=20):
    population = [[random.randint(0, 1) for _ in range(5)] for _ in range(pop_size)]
    
    for gen in range(generations):
        fitness_scores = [fitness_function(binary_to_int(c)) for c in population]
        
        best_idx = fitness_scores.index(max(fitness_scores))
        print(f"Generation {gen}: Best x = {binary_to_int(population[best_idx])}, Fitness = {max(fitness_scores)}")
        
        new_population = []
        for _ in range(pop_size):
            i, j = random.sample(range(pop_size), 2)
            winner = population[i] if fitness_scores[i] > fitness_scores[j] else population[j]
            new_population.append(winner[:])
            
        for i in range(0, pop_size, 2):
            if random.random() < 0.7:
                point = random.randint(1, 4)
                new_population[i][point:], new_population[i+1][point:] = new_population[i+1][point:], new_population[i][point:]
        
        for i in range(pop_size):
            if random.random() < 0.1:
                bit = random.randint(0, 4)
                new_population[i][bit] = 1 - new_population[i][bit]
        
        population = new_population

    final_fitness = [fitness_function(binary_to_int(c)) for c in population]
    best_final_idx = final_fitness.index(max(final_fitness))
    best_x = binary_to_int(population[best_final_idx])
    return best_x, max(final_fitness)

best_x, best_f = genetic_algorithm()
print(f"\nFinal optimal solution: x = {best_x}, f(x) = {best_f}")
```
