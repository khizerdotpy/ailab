```python
num=int(input("Enter number of students: "))

def studentrecord(num):
    student = []
    marks = []

    for i in range(num):
        stu=input("Enter student: ")
        student.append(stu)
        mk=int(input("Enter marks for that student: "))
        marks.append(mk)

        if marks[i] >=85:
            print("Student Name: ",student[i],"\nMarks: ",marks[i],"\nGrade: A")
        elif marks[i] >=70:
            print("Student Name: ",student[i],"\nMarks: ",marks[i],"\nGrade: B")
        elif marks[i] >=50:
            print("Student Name: ",student[i],"\nMarks: ",marks[i],"\nGrade: C")
        else:
            print("Student Name: ",student[i],"\nMarks: ",marks[i],"\nGrade: Fail")


studentrecord(num)
```
