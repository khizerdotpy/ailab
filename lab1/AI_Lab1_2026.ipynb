#**========================= Functions =========================**

```python
# Function definition is here
def non_paraFunc ():
  print("This is Non_parametric Function")

non_paraFunc()
```

```python
# Function definition is here
def paraFunc (a,b):
  """This is parametric Function"""
  #Takes the value of a and b parameters provided by the function user
  return a + b

paraFunc(1,2)
```

```python
# Function definition is here
def changeme( mylist ):
  '''This changes a passed list into this function'''
  print("Values inside the function before change: ", mylist)
  mylist[2]=50
  print ("Values inside the function after change: ", mylist)
  return

#Now you can call changeme function
mylist = [10,20,30]
changeme( mylist )
print ("Values outside the function: ", mylist)
```

#Return Statement

```python
# Function definition is here
def sum( arg1, arg2 ):
  # Add both the parameters and return them."
  total = arg1 + arg2
  print ("Inside the function : ", total)
  return total

# Now you can call sum function
total = sum( 10, 20 )
print ("Outside the function : ", total )
```

#No Return (NOne)

```python
def print_message(message):
    print('Print Statement : ', message)
    return  # This is equivalent to return None

result = print_message("Hello, World!")
#We cannot store data if nothing is returned from the statement
print(result)  # Output: None
```

#Retur Something

```python
def print_message(message):
    print('Print Statement : ', message)
    return message

result = print_message("Hello, World!")
#We cannot store data if soemthing is returned from the statement
print(result)  # Output: Message
```

#Fixe Parametric Function

```python
#Function with Fixed Parameters
def add(a, b):
  '''A and B are Fixed Parameters'''
  return a + b

add(1,2,3) #If we provide extra arguments, It will return TypeError
```

```python
#Function with Fixed Parameters
def display_names(names):
  '''This is Fixed Parametric Function'''
  return names

display_names('xyz')
```

#Functions types based on arguments

```python
#Positional Argument

def details(name,age):
  print('Name : ', name)
  print('Age : ', age)

details(77,'xyz')
```

```python
#Default Argument

def details(name,age=13):
  print('Name : ', name)
  print('Age : ', age)

details('xyz')
```

```python
#Keyword  Argument

def details(name,age,height):
  print('Names : ' ,name)
  print('age : ' ,age)
  print('height : ' ,height)

details(name= 'Ali', age= 99, height= 5.5)
```

#Variable Length Functions

```python
# Variable Length Args Argument

def details(*numbers):
  print('numbers : ' ,numbers)

details(1,2,3)
```

```python
#Keyword Variable Length Args Argument

def details(**name):
  print('Names : ' ,name)

details(n1= 'xyz', n2= 'ghf', n3= 'abc')
```
