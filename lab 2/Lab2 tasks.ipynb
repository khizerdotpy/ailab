# Task 1


```python
class Vehicle:

    def __init__(self,vehicle_id,brand,rent_per_day):
        self.vehicle_id=vehicle_id
        self.brand=brand
        self.rent_per_day=rent_per_day


    def display_details(self):
        print(f"Vehicle ID: {self.vehicle_id}")
        print(f"Brand: {self.brand}")
        print(f"Rent Per Day: ${self.rent_per_day}")


    def calculate_rent(self,days):
        totrent=self.rent_per_day*days

        print(f"Rent for Vehicle {self.vehicle_id} for {days} days is ${totrent}")



v1=Vehicle("PK3000","Koenigsegg",5000)
v2=Vehicle("PK1000","Toyota",700)


v1.display_details()
v2.display_details()

v1.calculate_rent(3)
```

# Task 2

```python
class Employee:

    def __init__(self,name,emp_id):
        self.name=name
        self.emp_id=emp_id

    def calculate_salary(self):
        #this is to ensure polymorphism as methods in base class are overridden in child classes. raises exception here as its not upposed to be implemented here
        raise NotImplementedError("Child Classes must implement this method")


class FullTimeEmployee(Employee):

    def __init__(self,name,emp_id,monthly_salary=0):
        super().__init__(name,emp_id)
        self.monthly_salary=monthly_salary

    def calculate_salary(self):
        return self.monthly_salary



class PartTimeEmployee(Employee):

    def __init__(self,name,emp_id,hours_worked=0,hourly_rate=0):
        super().__init__(name,emp_id)
        self.hours_worked=hours_worked
        self.hourly_rate=hourly_rate


    def calculate_salary(self):
        return self.hours_worked*self.hourly_rate





e1=FullTimeEmployee("Hassan",100,25000)
e2=PartTimeEmployee("Ali",200,120,25)


s1=e1.calculate_salary()
print(f"Employee 1: {e1.name} has a monthly salary of: {s1}")

s2=e2.calculate_salary()
print(f"Employee 2: {e2.name} has a salary of: {s2}")

        
```

```python
class Student:

    def __init__(self,name=" ",marks=-1):
        self._name=name
        self._marks=marks




    def set_marks(self,marks):
        self._marks=marks



    def get_marks(self):
        return self._marks

    def get_name(self):
        return self._name

    def calculate_grade(self,mark):

        if(mark>=80):
            print("A")
        elif (mark>=70):
            print("B")
        elif (mark>=60):
            print("C")
        else:
            print("F")




students=[]

for i in range(2):
    name=input(f"Enter name for student {i+1}: ")
    markin=int(input(f"Enter marks for student {i+1}: "))

    students.append(Student(name,markin))

print("\n")

for i in range(2):
    omarks=students[i].get_marks()
    oname=students[i].get_name()
    
    print(f"Marks for student {oname} are: {omarks}. Their grade is:",end=" ")
    students[i].calculate_grade(omarks)



```
