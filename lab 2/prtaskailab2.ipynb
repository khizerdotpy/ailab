```python
class lights:
  def __init__(self,roomn,st):
    self.room_name=roomn
    self.status=st

  def onfunc(self):
    status=True
    print(f"light for room: {self.room_name} turned {'ON' if self.status==True else 'OFF'}")


  def offfunc(self):
    status=False
    print(f"light for room: {self.room_name} turned {'OFF' if self.status==False else 'ON'}")

  def getStatus(self):
    print(f"Current Status is: {self.status}")






livingroom= lights("Living Room","OFF")
bedroom= lights("Bedroom","OFF")

livingroom.onfunc()
bedroom.offfunc()
livingroom.getStatus()
bedroom.getStatus()
```

# practice task 2

```python
class Staff:
    def __init__(self,nm,idnu,dpt):
        self.name=nm
        self.staff_id=idnu
        self.department=dpt

    def display_info(self):
        print(f"Name: {self.name}")
        print(f"Staff ID: {self.staff_id}")
        print(f"Department: {self.department}")


class Teacher(Staff):
    def __init__(self,nm,idnu,dpt,crs,sal):
        super().__init__(nm,idnu,dpt)
        self.course=crs
        self.salary=sal
    

    def teachcourse(self):
        print(f"{self.name} is currently teaching {self.course}")


    def display_info(self):
        super().display_info()
        print(f"Course: {self.course}")
        print(f"Salary: {self.salary}")

class Admin(Staff):
    def __init__(self,nm,idnu,dpt,rol,whours,task):
        super().__init__(nm,idnu,dpt)
        self.role=rol
        self.working_hours=whours
        self.tasks=task

    def display_info(self):
        super().display_info()
        print(f"Role: {self.role}")
        print(f"Working Hours: {self.working_hours}")
        print(f"Tasks: {self.tasks}")


    def perform_task(self):
        print(f"{self.name} is currently working on the task: {self.tasks}.")

class Research_Assistant(Staff):
    def __init__(self,nm,idnu,dpt,rtpc,st,wrk):
        super().__init__(nm,idnu,dpt)
        self.research_topic=rtpc
        self.stipend=st
        self.work_on_research=wrk
        
    def display_info(self):
        super().display_info()
        print(f"Research Topic: {self.research_topic}")
        print(f"Stipend: {self.stipend}")
        print(f"Works on Research: {self.work_on_research}")


    def currentwork(self):
        print(f"{self.name} is currently working on research in the field: {self.research_topic} regarding {self.work_on_research}.")


t1=Teacher("ali",100,"cs","math",7000)

a1=Admin("hamza",101,"cs","IT Programmer","10","Fix bugs")

ra1=Research_Assistant("ahmed",102,"biology","Marine Biology",5000,"Sharks")



print("Teacher Details")
t1.display_info()
t1.teachcourse()
    
print("\nAdmin Details")
a1.display_info()
a1.perform_task()
    
print("\nResearch Assistant Details")
ra1.display_info()
ra1.currentwork()







    
```

```python
class BankAccount:
    def __init__(self, name, account_number, balance = 0):
        self.name = name
        self.account_number = account_number
        self.__balance = balance
        
    def deposit(self, amount):
        if amount <= 0:
            print("Error: Deposit amount cannot be negative.")
        else:
            self.__balance += amount
            print(f"${amount} has been deposited to account number: {self.account_number}.\nNew Balance: ${self.__balance}")
            
    def withdraw(self, amount):
        if amount <= 0:
            print("Error: Withdrawal amount cannot be negative.")
        elif self.__balance < amount:
            print("Error: Insufficient balance!")
        else:
            self.__balance -= amount
            print(f"${amount} has been withdrawn from account number: {self.account_number}.\nNew balance: ${self.__balance}")
            
    def get_balance(self):
        return self.__balance





account1 = BankAccount("ali", "PK3300", 10000)
account2 = BankAccount("hamza", "PK5600", 3000)
account3 = BankAccount("john cena", "PK3980", 9000)

account1.deposit(200)
account1.withdraw(1500)
print()

account2.deposit(1000)
account2.withdraw(15000)
print()

account3.withdraw(2500)
account3.deposit(7000)

print("\nAccount 1")
print(f"User Name: {account1.name}")
print(f"Account Number: {account1.account_number}")
print(f"Balance: ${account1.get_balance()}")

print("\nAccount 2")
print(f"User Name: {account2.name}")
print(f"Account Number: {account2.account_number}")
print(f"Balance: ${account2.get_balance()}")

print("\nAccount 3")
print(f"User Name: {account3.name}")
print(f"Account Number: {account3.account_number}")
print(f"Balance: ${account3.get_balance()}")
```
