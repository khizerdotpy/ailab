```python

class Environment:
  def __init__(self,cleanliness):
    self.cleanliness=cleanliness



  def get_percept(self):
    return self.cleanliness


  def clean_room(self):
    self.cleanliness='Clean'


class ReflexAgent:
  def __init__(self):
    pass


  def action(self,percept):
    if percept=='Clean':
      return "Room is clean. No action."
    elif percept=='Dirty':
      return "Room is not clean. Cleaning."
    else:
      return "Invalid percept input"

def run_agent(agent,environment):
  percept=environment.get_percept()
  act=agent.action(percept)

  print(f"Percept: {percept}, Action: {act}")

agent=ReflexAgent()
room=Environment(input("Enter state(clean,dirty):"))

def run_agentstepwise(agent,environment,steps):
  for step in range(steps):
      percept=environment.get_percept()
      act=agent.action(percept)

      print(f"Percept: {percept}, Action: {act}")
      if(percept=='Dirty'):
        environment.clean_room()



run_agent(agent,room)
print("\n")
run_agentstepwise(agent,room,5)
```
