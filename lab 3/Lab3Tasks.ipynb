# Lab Task 2

```python

import random


class ClassEnvironment:
    def __init__(self):
        self.student_present=random.choice([True,False])
        self.light_status = "Off"

    def get_percept(self):
        return {'student':self.student_present,'light':self.light_status}

    def light_on(self):
        self.light_status="On"
    def light_off(self):
        self.light_status="Off"

    def update_presence(self):
        self.student_present=random.choice([True,False])

        
class LightAgent:

    def __init__(self):
        self.model = {'prev_students': None, 'prev_light': None, 'curr_students': None, 'curr_light': None}

    def action(self,percept):
        self.model['prev_students']=self.model['curr_students']
        self.model['prev_light']=self.model['curr_light']
        self.model['curr_students']=percept['student']
        self.model['curr_light']=percept['light']
        
        if self.model['curr_students'] == True and self.model['curr_light'] == "Off":
            return "Turn light ON"
        elif self.model['curr_students'] == False and self.model['curr_light'] == "On":
            return "Turn light OFF"
        else:
            return "No action"





def run_LightAgent(environment,agent,steps):

    for step in range(steps):
        print(f"\nStep {step+1}:",end=" ")

        environment.update_presence()
        percept=environment.get_percept()
        act=agent.action(percept)


        if act=="Turn light ON":
            environment.light_on()
        elif act=="Turn light OFF":
            environment.light_off()

        print(f"Percept: {percept}, Action: {act}, Agent Model: {agent.model}")
        





agent=LightAgent()
environment=ClassEnvironment()
steps=8
run_LightAgent(environment,agent,steps)
```

# Task 3

```python

class GoalBasedAgent:

    def __init__(self):
        self.goal = "Complete Subjects"

    def formulate_goal(self, percept):
        if percept is not None:
            self.goal = "Study"
        else:
            self.goal = "Done"

    def act(self, percept):
        self.formulate_goal(percept)

        if self.goal == "Study":
            return f"Studying {percept}"
        else:
            return "Goal Achieved: All subjects completed"


class StudyEnvironment:

    def __init__(self):
        self.subjects = ["AI", "Math", "Physics"]

    def get_percept(self):
        if self.subjects:
            return self.subjects[0]   # next subject
        return None

    def complete_subject(self):
        if self.subjects:
            self.subjects.pop(0)


def run_agent(agent, environment):

    step = 1
    while True:
        percept = environment.get_percept()
        action = agent.act(percept)

        print(f"Step {step}: Percept - {percept}, Action - {action}")

        if percept is not None:
            environment.complete_subject()
        else:
            break

        step += 1

agent = GoalBasedAgent()
environment = StudyEnvironment()

run_agent(agent, environment)
```

# Task 4

```python
class RestaurantEnvironment:

    def __init__(self):
        self.restaurants = {
            "A": {"distance": 3, "rating": 7},
            "B": {"distance": 5, "rating": 9}
        }

    def get_data(self):
        return self.restaurants

class UtilityAgent:

    def action(self, data):

        best = None
        best_utility = -99999

        for name, info in data.items():
            utility = info["rating"] - info["distance"]
            print(f"Restaurant {name} Utility = {utility}")

            if utility > best_utility:
                best_utility = utility
                best = name

        print(f"\nSelected Restaurant: {best}")


def run_utility_agent(env, agent):
    agent.action(env.get_data())


env = RestaurantEnvironment()
agent = UtilityAgent()
run_utility_agent(env, agent)
```

# Task 5

```python
import random

class GameEnvironment:

    def __init__(self):
        pass

    def get_reward(self, action):
        if action == "Play":
            return 5
        return 1


class LearningAgent:

    def __init__(self):
        self.Q = {"Play": 0, "Rest": 0}
        self.alpha = 0.1

    def choose_action(self):
        return random.choice(["Play", "Rest"])

    def action(self, env, step):
        act = self.choose_action()
        reward = env.get_reward(act)

        self.Q[act] += self.alpha * (reward - self.Q[act])

        print(f"Step {step}: Action {act} , Reward {reward}")


def run_learning_agent(env, agent):
    for step in range(1, 11):
        agent.action(env, step)

    print("Q-table Updated:", agent.Q)


env = GameEnvironment()
agent = LearningAgent()
run_learning_agent(env, agent)
```

# Task 6

```python
class FireEnvironment:

    def __init__(self):
        self.rooms = {
            'a': 'safe','b': 'safe','c': 'fire',
            'd': 'safe','e': 'fire','f': 'safe',
            'g': 'safe','h': 'safe','j': 'fire'
        }

    def get_status(self, room):
        return self.rooms[room]

    def extinguish(self, room):
        self.rooms[room] = 'safe'

    def display(self):
        order = ['a','b','c','d','e','f','g','h','j']
        print("\nEnvironment:")
        for i in range(0, 9, 3):
            row = ""
            for r in order[i:i+3]:
                row += "游댠 " if self.rooms[r] == 'fire' else "游릴"
            print(row)

class FireRobot:

    def __init__(self):
        self.path = ['a','b','c','d','e','f','g','h','j']

    def action(self, env):
        for room in self.path:
            print(f"\nMoved to {room}")

            if env.get_status(room) == 'fire':
                print("Fire detected! Extinguishing...")
                env.extinguish(room)
            else:
                print("Room safe")

            env.display()




env = FireEnvironment()
robot = FireRobot()
robot.action(env)

print("\nFinal Status: All rooms safe")
```
