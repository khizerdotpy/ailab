# Question 1

```python
num=int(input("Enter number of students: "))

def studentrecord(num):
    student = []
    marks = []

    for i in range(num):
        stu=input("Enter student: ")
        student.append(stu)
        mk=int(input("Enter marks for that student: "))
        marks.append(mk)

        if marks[i] >=85:
            print("Student Name: ",student[i],"\nMarks: ",marks[i],"\nGrade: A")
        elif marks[i] >=70:
            print("Student Name: ",student[i],"\nMarks: ",marks[i],"\nGrade: B")
        elif marks[i] >=50:
            print("Student Name: ",student[i],"\nMarks: ",marks[i],"\nGrade: C")
        else:
            print("Student Name: ",student[i],"\nMarks: ",marks[i],"\nGrade: Fail")


studentrecord(num)
```

# Question 2

```python
enum=int(input("Enter number: "))

def even(enum):
    even=2
    count=0
    fl=True
    print("Even numbers: ", end=" ")
    while fl:
        if (even>enum):
            fl=False
        else:
            print(even,end= " ")
            even=even+2
            count=count+1


    print(f"\nTotal even numbers: {count}")
        


even(enum)


```

# Question 3

```python
unum=int(input("Enter how many students: "))

def record(unum):
  udict = {}

  for i in range(unum):
    key=input("Enter student name: ")
    value=int(input("Enter student marks: "))

    udict[key]=value


  print("Student Records: ")

  for key, value in udict.items():
    print(f"{key} : {value}")



record(unum)
```
